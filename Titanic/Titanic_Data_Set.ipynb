```python
#15. What is the survival rate for each passenger class?
print(titanic.groupby('Pclass')['Survived'].mean().round(2))
pd.crosstab(titanic['Pclass'], titanic['Survived'], normalize='index')
```

```python
# 16. Sort passengers by fare in descending order
titanic.sort_values(by='Fare',ascending=False)[['PassengerId','Name']]
```

```python
#17. Replace all "male"/"female" in the Sex column with "M"/"F".
titanic.loc[titanic['Sex'] == 'male', 'Sex'] = 'M'
titanic.loc[titanic['Sex'] == 'female', 'Sex'] = 'F'
titanic['Sex'] = titanic['Sex'].replace({'male': 'M', 'female': 'F'})
```

```python
titanic
```

```python
#18. Rename SibSp to Siblings_Spouses and Parch to Parents_Children
titanic.rename(columns={'SibSp':'Siblings_Spouses','Parch':'Parents_Children'},inplace=True)
```

```python
#19. Create a new column FamilySize = Siblings_Spouses + Parents_Children
titanic['FamilySize']=titanic['Siblings_Spouses']+titanic['Parents_Children']
```

```python
#20. Create a column IsMinor which is True if age < 18
titanic['IsMinor']=np.where(titanic['Age']<18,True,False)
titanic['IsMinor'] = titanic['Age'] < 18
titanic['IsMinor'] = titanic['Age'].apply(lambda x: True if x < 18 else False)
```

```python
titanic[titanic['Age']<18][['Age','IsMinor']].head(10)
```

```python
# 21. Create a new column FareBin based on fare ranges:
# 0–10 → 'Low'

# 10–50 → 'Medium'

# 50+ → 'High'
titanic['FareBin']=pd.cut(titanic['Fare'],bins=[0,10,50,titanic['Fare'].max()],labels=['low','medium','high'])
```

```python
titanic[titanic['Age']==titanic['Age'].min()][['Name','Pclass']]
```

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
```

```python
titanic['Embarked'].isna().sum()
```

```python
# 24.Drop rows where Embarked is null
titanic = titanic.dropna(subset=['Embarked'])
```

```python
# 27.Find the survival rate for passengers with fare above average.
avg_fare=titanic['Fare'].mean()
high_fare_passengers=titanic[titanic['Fare']>avg_fare]
survival_rate=high_fare_passengers['Survived'].mean().round(2)
survival_rate
```

```python
# 28.	Who had the highest fare? Did they survive?
titanic[titanic['Fare']==titanic['Fare'].max()][['Name', 'Fare', 'Survived', 'Pclass']]
```

```python
# 29: Which class had the most children on board?
children = titanic[titanic['Age'] < 18]
children['Pclass'].value_counts()
```

We’ll use Matplotlib and Seaborn to visualize key Titanic insights.

🚢 Titanic Visualization Plan:
1. Survival Count
Bar chart of survivors vs non-survivors

2. Survival by Gender
Grouped bar or countplot

3. Survival by Class
Bar plot

4. Age Distribution
Histogram or KDE plot

5. Fare Distribution
Histogram or boxplot

6. Heatmap of Missing Values
Use sns.heatmap

7. Survival by Class & Gender
Stacked/grouped bar using Seaborn

8. Violin/Box Plot of Age vs Class/Survival
9. Correlation Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
```

```python
# 1. Survival Count
# Count survivors and non-survivors
survival_counts = titanic['Survived'].value_counts()

# Define labels and values
labels = ['Did not survive', 'Survived']
values = [survival_counts[0], survival_counts[1]]

# Plot using plt.bar
plt.bar(labels, values, color=['red', 'green'])
plt.title('Survival Count')
plt.xlabel('Survival Status')
plt.ylabel('Passenger Count')
plt.show()
```

```python
#  2. Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=titanic)
plt.title('Survival by Gender')
plt.xticks([0, 1], ['Female', 'Male'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
#  3. Survival by Class
sns.countplot(x='Pclass', hue='Survived', data=titanic)
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

```

```python
plt.figure(figsize=(10, 5))
sns.histplot(titanic['Age'],bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(titanic['Fare'], bins=40, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=titanic)
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
